```python
import numpy as np
import pandas as pd
from datetime import date, datetime
import datetime as dt
import matplotlib.pyplot as plt
```

```python
Sales2020 = pd.read_csv('Historical_sales2020.csv')
Sales2021 = pd.read_csv('Historical_sales2021.csv')
Sales2022 = pd.read_csv('Historical_sales2022.csv')
```

```python
Sales2021['unit_price'] = Sales2021['unit_price'].str.replace(',','.').astype(float)
Sales2020['unit_price'] = Sales2020['unit_price'].str.replace(',','.').astype(float)
```

```python
Sales20_22 = pd.concat([Sales2020, Sales2021, Sales2022], axis = 0)
Sales20_22
```

```python
Sales20_22['transaction_date'] = pd.to_datetime(Sales20_22['transaction_date'])
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.boxplot([Sales20_22['unit_price'], Sales20_22['quantity_sold']])
plt.show()
```

```python
Q1 = np.percentile(Sales20_22['unit_price'], 25, method='midpoint')
Q3 = np.percentile(Sales20_22['unit_price'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
upper = Q3+1.5*IQR
upper_array = np.array(Sales20_22['unit_price'] >= upper)
print("Upper Bound:", upper)
print(upper_array.sum())
 
lower = Q1-1.5*IQR
lower_array = np.array(Sales20_22['unit_price'] <= lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
Sales20_22 = Sales20_22[(Sales20_22["unit_price"] > lower) & (Sales20_22["unit_price"] < upper)]
Sales20_22
```

```python
Q1 = np.percentile(Sales20_22['quantity_sold'], 25, method='midpoint')
Q3 = np.percentile(Sales20_22['quantity_sold'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
upper = Q3+1.5*IQR
upper_array = np.array(Sales20_22['quantity_sold'] >= upper)
print("Upper Bound:", upper)
print(upper_array.sum())
 
lower = Q1-1.5*IQR
lower_array = np.array(Sales20_22['quantity_sold'] <= lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
Sales20_22 = Sales20_22[(Sales20_22["quantity_sold"] > lower) & (Sales20_22["quantity_sold"] < upper)]
Sales20_22
```

```python
Sales20_22.to_csv('Historical_sales.csv', index=False)
```
