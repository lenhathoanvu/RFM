##### Câu 4

Một trong những chiến lược trọng điểm của Meki Group trong thời gian tới là cải thiện và nâng cao trải nghiệm mua sắm của khách hàng, đẩy mạnh số lượng khách hàng mới và tăng tỷ lệ giữ chân khách hàng cũ. Do đó, công ty muốn bạn sử dụng kỹ năng phân tích để tính toán và phân chia các tệp khách hàng mới và khách hàng trung thành thông qua lịch sử mua hàng trong bộ dataset đã cung cấp

Để phân chia các khách hàng mới và khách hàng trung thành dựa trên lịch sử mua hàng, mình sử dụng phân tích RFM (Recency, Frequency, Monetary). Phân tích RFM là một phương pháp dựa trên hành vi để nhóm khách hàng vào các phân khúc. Nó nhóm các khách hàng dựa trên các tiêu chí sau:

- Recency: là khoảng thời gian kể từ lần mua hàng gần nhất của khách hàng. Khách hàng có recency nhỏ thường có xu hướng mua hàng nhiều hơn và là khách hàng trung thành.
- Frequency: là số lần mua hàng của khách hàng trong một khoảng thời gian nhất định. Khách hàng có frequency cao thường có nhu cầu mua hàng lớn và là khách hàng tiềm năng.
- Monetary: là tổng số tiền mà khách hàng đã chi tiêu cho các sản phẩm trong một khoảng thời gian nhất định. Khách hàng có monetary cao thường có giá trị cao và là khách hàng quan trọng.

## Recency

```python
Recency = Customer_22.groupby(by='customer_id',
                        as_index=False)['transaction_date'].max()
Recency.columns = ['Customer_id', 'Last_Purchase_Date']
Recency["Last_Purchase_Date"] = pd.to_datetime(Recency["Last_Purchase_Date"])
recent_date = Recency['Last_Purchase_Date'].max()
Recency['Recency'] = Recency['Last_Purchase_Date'].apply(
    lambda x: (recent_date - x).days)
```

```python
Recency.head(5)
```

## Frequency

```python
Frequency_22 = Customer_22.drop_duplicates().groupby(
    by=['customer_id'], as_index=False)['transaction_date'].count()
Frequency_22.columns = ['Customer_id', 'Frequency']

Frequency_21 = Customer_21.drop_duplicates().groupby(
    by=['customer_id'], as_index=False)['transaction_date'].count()
Frequency_21.columns = ['Customer_id', 'Frequency']

Frequency_20 = Customer_20.drop_duplicates().groupby(
    by=['customer_id'], as_index=False)['transaction_date'].count()
Frequency_20.columns = ['Customer_id', 'Frequency']
```

```python
Frequency_22.head()
```

```python
Frequency_21.head()
```

```python
Frequency_20.head()
```

```python
Frequency = Frequency_22.merge(Frequency_21, on = "Customer_id", how = "outer")
Frequency = Frequency.merge(Frequency_20, on = "Customer_id", how = "outer")
```

```python
Frequency.columns = ["Customer_id", "Frequency_22", 
                     "Frequency_21", "Frequency_20"]
Frequency["Frequency"] = Frequency["Frequency_22"] + Frequency["Frequency_21"] 
+ Frequency["Frequency_21"]
```

```python
Frequency.drop(columns=["Frequency_22", "Frequency_21", "Frequency_20"], 
               inplace=True)
```

```python
Frequency.head(5)
```

## Monetary

```python
Customer_21["unit_price"] = Customer_21["unit_price"].str.replace(",", ".").astype(float)
Customer_21
```

```python
Customer_20["unit_price"] = Customer_20["unit_price"].str.replace(",", ".").astype(float)
Customer_20
```

```python
Customer_22["revenue"] = Customer_22["quantity_sold"]*Customer_22["unit_price"]
Customer_21["revenue"] = Customer_21["quantity_sold"]*Customer_21["unit_price"]
Customer_20["revenue"] = Customer_20["quantity_sold"]*Customer_20["unit_price"]
```

```python
Monetary_22 = Customer_22.groupby(by='customer_id', as_index=False)['revenue'].sum()
Monetary_22.columns = ['Customer_id', 'Monetary']
Monetary_22.head()
```

```python
Monetary_21 = Customer_21.groupby(by='customer_id', as_index=False)['revenue'].sum()
Monetary_21.columns = ['Customer_id', 'Monetary']
Monetary_21.head()
```

```python
Monetary_20 = Customer_20.groupby(by='customer_id', as_index=False)['revenue'].sum()
Monetary_20.columns = ['Customer_id', 'Monetary']
Monetary_20.head()
```

```python
Monetary = Monetary_22.merge(Monetary_21, on = "Customer_id", how = "outer")
Monetary = Monetary.merge(Monetary_20, on = "Customer_id", how = "outer")
Monetary
```

```python
Monetary.columns = ["Customer_id", "Monetary_22", 
                     "Monetary_21", "Monetary_20"]
Monetary["Monetary"] = Monetary["Monetary_22"] + Monetary["Monetary_21"] 
+ Monetary["Monetary_20"]
```

```python
Monetary.drop(columns=["Monetary_22", "Monetary_21", "Monetary_20"], 
               inplace=True)
```

```python
Monetary
```

## Merging all three columns in one dataframe

```python
RF = Recency.merge(Frequency, on='Customer_id')
RFM = RF.merge(Monetary, on='Customer_id').drop(
    columns='Last_Purchase_Date')
RFM.head(5)
```

## Ranking Customer’s based upon their recency, frequency, and monetary score

```python
RFM['R_rank'] = RFM['Recency'].rank(ascending=False)
RFM['F_rank'] = RFM['Frequency'].rank(ascending=True)
RFM['M_rank'] = RFM['Monetary'].rank(ascending=True)
 
# normalizing the rank of the customers
RFM['R_rank_norm'] = (RFM['R_rank']/RFM['R_rank'].max())*100
RFM['F_rank_norm'] = (RFM['F_rank']/RFM['F_rank'].max())*100
RFM['M_rank_norm'] = (RFM['F_rank']/RFM['M_rank'].max())*100
 
RFM.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
 
RFM.head()
```

## Calculating RFM score

```python
RFM['RFM_Score'] = 0.15*RFM['R_rank_norm']+0.28 * \
    RFM['F_rank_norm']+0.57*RFM['M_rank_norm']
RFM['RFM_Score'] *= 0.05
RFM = RFM.round(2)
RFM[['Customer_id', 'RFM_Score']].head(7)
```

## Rating Customer based upon the RFM score

- RFM score > 4.5 : Top Customer
- 4.5 > RFM score > 4 : High Value Customer
- 4 > RFM score >3 : Medium value customer
- 3 > RFM score > 1.6 : Low-value customer
- RFM score < 1.6 : Lost Customer

```python
RFM["Customer_segment"] = np.where(RFM['RFM_Score'] > 4.5, "Top Customers",(np.where(
    RFM['RFM_Score'] > 4,"High value Customer",(np.where(RFM['RFM_Score'] > 3,"Medium Value Customer",
                                                            np.where(RFM['RFM_Score'] > 1.6,
                                                                     'Low Value Customers', 'Lost Customers'))))))
RFM[['Customer_id', 'RFM_Score', 'Customer_segment']].head(20)
```

## Visualizing the customer segments

```python
plt.pie(RFM.Customer_segment.value_counts(),
        labels=RFM.Customer_segment.value_counts().index,
        autopct='%.0f%%')
plt.show()
```
