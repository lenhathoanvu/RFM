```python
import numpy as np
import pandas as pd
from datetime import date, datetime
import datetime as dt
import matplotlib.pyplot as plt
```

```python
Sales2020 = pd.read_csv('Historical_sales2020.csv')
Sales2021 = pd.read_csv('Historical_sales2021.csv')
Sales2022 = pd.read_csv('Historical_sales2022.csv')
```

```python
Sales2021['unit_price'] = Sales2021['unit_price'].str.replace(',','.').astype(float)
Sales2020['unit_price'] = Sales2020['unit_price'].str.replace(',','.').astype(float)
```

```python
Sales20_22 = pd.concat([Sales2020, Sales2021, Sales2022], ignore_index=True)
Sales20_22
```

```python
Sales20_22['transaction_date'] = pd.to_datetime(Sales20_22['transaction_date'])
```

```python
Sales20_22.to_csv('Historical_sales.csv', index=False)
```
